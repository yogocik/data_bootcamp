# Pandas Library Usage

```python
import pandas as pd
```

```python
product_data = pd.read_csv('./products.csv')
product_data.head()
```

# Select columns or attributes of table

```python
product_data[['id','cost']]
```

```python
# By Index
product_data.loc[0]
```

# Filter the data based on certain values or conditions

```python
selected = product_data.query("cost > 100")
selected
```

```python
selected = product_data.query("cost > 100").copy()
selected
```

```python
selected = product_data[(product_data['cost'] > 100) & (product_data['department'] == 'Men')]
selected
```

# Update table values

```python
product_data.head()
```

```python
product_data.loc[0,['cost','brand']] = [1000, 'Jeans']
product_data.loc[0]
```

```python
baris = [0,1,2]
data = [2000, 5000, 10000]
for index in range(len(baris)):
    product_data.loc[baris[index],['cost']] = [data[index]]
product_data.head()
```

```python
baris = [0,1,2]
list(range(len(baris)))
```

```python
product_data.head()
```

```python
def multiply(number:float) -> float:
    return 1.5 * number

product_data['new_cost'] = product_data['cost'].map(multiply)
product_data.head()
```

```python
product_data['discounted_cost'] = product_data['cost'].map(lambda x: x * (1-0.1))
product_data.head()
```

```python
new_product_data = product_data[['id', 'cost','discounted_cost','brand','category']]
new_product_data
```

# Export the data into certain format

```python
# To Excel (Need to install additional module, "openpyxl")
product_data.to_excel(excel_writer="./updated_products.xlsx", index=False)
```

```python
# To Excel (Need to install additional module, "openpyxl")
product_data.to_excel(excel_writer="./updated_products_general.xlsx", index=False, float_format="%g")
```

```python
# To CSV
product_data.to_csv(path_or_buf="./updated_products.csv", index=False)
```

```python
# To CSV
product_data.to_csv(path_or_buf="./updated_products_general.csv", index=False, float_format="%g")
```
